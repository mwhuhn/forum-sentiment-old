## Import

```python
from lemmatize import *
import pickle
import pandas as pd
```

## Get Frequency Distributions

```python
fd_sn = get_freq_dist("special-needs", "all")
```

```python
fd_td = get_freq_dist("toddler", "all")
```

```python
fd_sc = get_freq_dist("school", "all")
```

## Compare Frequencies

```python
wf_sn = pd.DataFrame.from_dict(fd_sn, orient='index', columns=['sn'])
wf_td = pd.DataFrame.from_dict(fd_td, orient='index', columns=['td'])
wf_sc = pd.DataFrame.from_dict(fd_sc, orient='index', columns=['sc'])
wf = wf_sn.join(wf_td, how='outer')
wf = wf.join(wf_sc, how='outer')
wf.fillna(0, inplace=True)
```

```python
wf.head()
```

```python
wf['sn_rate'] = wf['sn'] /  wf['sn'].sum()
wf['td_rate'] = wf['td'] /  wf['td'].sum()
wf['sc_rate'] = wf['sc'] /  wf['sc'].sum()
```

### Top N Words

```python
def print_top_n(df, col, n):
    i = list(df.columns).index(col)
    return df.sort_values(by=col, ascending=False).iloc[0:n]
```

```python
print_top_n(wf, 'sn_rate', 20)
```

```python
print_top_n(wf, 'td_rate', 20)
```

```python
print_top_n(wf, 'sc_rate', 20)
```

### Differences in Frequencies

```python
wf['sn_td_rate'] = wf['sn_rate'] / wf['td_rate']
wf['td_sn_rate'] = wf['td_rate'] / wf['sn_rate']
wf['sn_sc_rate'] = wf['sn_rate'] / wf['sc_rate']
wf['sc_sn_rate'] = wf['sc_rate'] / wf['sn_rate']
wf['sc_td_rate'] = wf['sc_rate'] / wf['td_rate']
wf['td_sc_rate'] = wf['td_rate'] / wf['sc_rate']
```

```python
min_rate = 0.00001
```

```python
mask = (wf['sn_rate'] > min_rate) & (wf['td_rate'] > min_rate)
wf[mask].sort_values(by='sn_td_rate', ascending=False).iloc[0:20]
```

```python
wf[mask].sort_values(by='td_sn_rate', ascending=False).iloc[0:20]
```

```python
mask = (wf['sn_rate'] > min_rate) & (wf['sc_rate'] > min_rate)
wf[mask].sort_values(by='sn_sc_rate', ascending=False).iloc[0:20]
```

```python
wf[mask].sort_values(by='sc_sn_rate', ascending=False).iloc[0:20]
```

```python
mask = (wf['td_rate'] > min_rate) & (wf['sc_rate'] > min_rate)
wf[mask].sort_values(by='td_sc_rate', ascending=False).iloc[0:20]
```

```python
wf[mask].sort_values(by='sc_td_rate', ascending=False).iloc[0:20]
```

## Spearman's Rank Correlation

```python
from scipy.stats import spearmanr
```

```python
wf['sn_rank'] = wf['sn_rate'].rank(ascending=False)
wf['td_rank'] = wf['td_rate'].rank(ascending=False)
wf['sc_rank'] = wf['sc_rate'].rank(ascending=False)
```

```python
spearmanr(wf['sn_rank'], wf['td_rank'])
```

```python
spearmanr(wf['sn_rank'], wf['sc_rank'])
```

```python
spearmanr(wf['td_rank'], wf['sc_rank'])
```
